```
#!/usr/bin/env python

# make sure to install these packages before running:
# pip install pandas
# pip install sodapy

import pandas as pd
from sodapy import Socrata

# Unauthenticated client only works with public data sets. Note 'None'
# in place of application token, and no username or password:
client = Socrata("chronicdata.cdc.gov", None)

# Example authenticated client (needed for non-public datasets):
# client = Socrata(chronicdata.cdc.gov,
#                  MyAppToken,
#                  userame="user@example.com",
#                  password="AFakePassword")

# First 2000 results, returned as JSON from API / converted to Python list of
# dictionaries by sodapy.
results = client.get_all("f8ti-h92k")

# Convert to pandas DataFrame
dbt_cdi = pd.DataFrame.from_records(results)
```

```
dvtid = pd.Series(dbt_cdi["datavaluetypeid"])
pd.unique(dvtid)
```

```
dia_df = dbt_cdi.filter(["datavaluetypeid", "datavaluetype", "yearstart","yearend" ,"datavalue","question", "stratificationcategoryid1", "stratification1"])
dia_df.describe()
```

```
dtvidlist = ['AGEADJRATE', 'NMBR', 'CRDRATE', 'PREV', 'CRDPREV', 'AGEADJPREV']
```

```
dia_dfl = []
for dtvid in dtvidlist:
  dia_dfl.append(dia_df[dia_df.datavaluetypeid == dtvid])
```

```
dia_dfl[0].head()
```

```
strats = ['RACE', 'GENDER', 'OVERALL']
dia_strat = []
for strat in strats:
  for i in range(len(dia_dfl)):
    dia_strat.append(dia_dfl[i][dia_dfl[i].stratificationcategoryid1 == strat])
```
